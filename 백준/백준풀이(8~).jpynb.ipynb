```python
# 2745번
# 진법변환 

"""
문제: B진법 수 N이 주어진다. 이 수를 10진법으로 바꿔 출력하는 프로그램을 작성하시오.
0진법을 넘어가는 진법은 숫자로 표시할 수 없는 자리가 있다. 이런 경우에는 다음과 같이 알파벳 대문자를 사용한다.

"""

n, b = input().split()  # 두개의 값을 입력받아 공백기준으로 분리 
b = int(b)              # 두 번째  입력값 b를 정수형으로 변환 
print(int(n,b))         # 첫 번째 입력값 n을 b 진수로 --> 10진수 변환 
```

```python
# 11005번
# 진법 변환2
"""
문제: 10진법 수 N이 주어진다. 이 수를 B진법으로 바꿔 출력하는 프로그램을 작성하시오.

10진법을 넘어가는 진법은 숫자로 표시할 수 없는 자리가 있다. 이런 경우에는 다음과 같이 알파벳 대문자를 사용한다.

A: 10, B: 11, ..., F: 15, ..., Y: 34, Z: 35

"""

num_list = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'  # 진수 변환시 사용할 문자 리스트

n, b = map(int, input().split())
answer = "" # 변환 결과를 저장할 빈 문자열 

while n != 0:     # n을 b로 나눈 나머지를 이용하여 해당 문자를 결과 문자열 앞에 추가 
    answer = num_list[n % b] + answer  # num_list에 찾아 해당 문자열 answer앞에 추가 
    n // b # n을 b로 나눈 몫으로 갱신 
print(answer) 
```
