```python
# 킹, 퀸, 룩, 비숍, 나이트, 폰
# 3003번 

"""
검정색 피스, 흰색 피스는 개수 올바르지x
체스는 총 16개의 피스를 사용, 킹1개, 퀸1개, 룩2개, 비숍2개, 나이트2개, 폰8개 구성 
동현이가 발견한 흰색 피스의 개수가 주여졌을 때, 몇 개를 더하거나 빼야 올바른 세트가 되는지 프로그램을 작성하시오  
"""

hint = [1,1,2,2,2,8]
DH = list(map(int,input().split()))

answer = []
for i in range(6):
    answer.append(hint[i]-DH[i])   # hint, dh리스트의 대응하는 요소들을 빼서 answer리스트에 추가  
    # 올바른 세트값을 창출 
for i in range(6):
    print(answer[i], end='') 
    
```

```python
# 별찍기 
# 2444번
"""
예제를 보고 규칙을 유추한 뒤에 별을 찍어보세요 
"""
n = int(input())
for i in range(1,n+1):
    print(" "*(n-i) + "*"*(2*i-1))
for i in range(n-1, 0, -1):
    print(" "*(n-i) + "*"*(2*i-1)) 
    
```

```python
# 팰린드룸인지 확인하기 
# 10988번 
"""
알파벳 소문자로만 이루어진 단어가 주어진다. 이때 이 단어가 팰린드롬인지 아닌지 확인하는 프로그램을 작성하시오. 
level, noon은 팰린드롬이고, baekjoon, online, judge는 팰린드롬이 아니다.
첫째 줄에 팰린드롬이면 1, 아니면 0을 출력한다. 
"""

word = list(str(input()))

if list(reversed(word)) == word:  
    print(1)
else:
    print(0)
```

```python
# 크로아티아 알파벳 
# 2941번
"""
예를 들어, ljes=njak은 크로아티아 알파벳 6개(lj, e, š, nj, a, k)로 이루어져 있다. 단어가 주어졌을 때, 몇 개의 크로아티아 알파벳으로 이루어져 있는지 출력한다.

dž는 무조건 하나의 알파벳으로 쓰이고, d와 ž가 분리된 것으로 보지 않는다. lj와 nj도 마찬가지이다. 위 목록에 없는 알파벳은 한 글자씩 센다.
"""

croatia = ["c=","c-","dz=","d-","lj","nj","s=","z="]

alpha = input() 
for i in croatia:
    alpha = alpha.replace(i,'*')  # croatia에 대한 알파벳을 *로 대체 따라서 목록에 알파벳은 한글짜로 센다. 

print(len(alpha)) 
```

```python
# 평점
# 25206번 
"""
인하대학교 컴퓨터공학과를 졸업하기 위해서는, 전공평점이 3.3 이상이거나 졸업고사를 통과해야 한다. 그런데 아뿔싸, 치훈이는 깜빡하고 졸업고사를 응시하지 않았다는 사실을 깨달았다!

치훈이의 전공평점을 계산해주는 프로그램을 작성해보자.

전공평점은 전공과목별 (학점 × 과목평점)의 합을 학점의 총합으로 나눈 값이다.

인하대학교 컴퓨터공학과의 등급에 따른 과목평점은 다음 표와 같다.

A+	4.5
A0	4.0
B+	3.5
B0	3.0
C+	2.5
C0	2.0
D+	1.5
D0	1.0
F	0.0
P/F 과목의 경우 등급이 P또는 F로 표시되는데, 등급이 P인 과목은 계산에서 제외해야 한다.

"""
grade_list = ["A+", "A0", "B+", "B0", "C+", "C0", "D+", "D0", "F"]
score_list = [4.5, 4.0, 3.5, 3.0, 2.5, 2.0, 1.5, 1.0, 0.0]

tot_credit = 0
tot_Score = 0

for i in range(20):
    a = list(map(str, input().split()))
    credit = float(a[i])
    if a[2] == "P":
        continue
    else: 
        idx = grade_list.index(a[2])
        score = score_list[idx]
        tot_score += (credit*score)
        tot_credit += credit
avg_score = float(tot_score / tot_credit)
print(avg_score) 
```
